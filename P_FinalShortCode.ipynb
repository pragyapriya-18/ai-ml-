**1.MNIST:--**

---

```python
#Importing
import numpy as np
import pandas as pd
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense

# Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Preprocess
x_train = x_train.reshape(60000, 28*28) / 255
x_test  = x_test.reshape(10000, 28*28) / 255

# Model
model = Sequential()
model.add(Dense(32, input_dim=28*28, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))


# Train
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=10, batch_size=128)

# Test accuracy
loss, acc = model.evaluate(x_test, y_test)
print("Test Accuracy:", acc)

# Optional: Predict one sample
y_pred = model.predict(x_test)
print("Predicted digit:", np.argmax(y_pred[0]))
```

```python
# 1️⃣ Import
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense

# 2️⃣ Load
iris = load_iris()
X = iris.data
y = to_categorical(iris.target, num_classes=3)

# 3️⃣ Prepare
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 4️⃣ Model
model = Sequential()
model.add(Dense(10, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))

# 5️⃣ Train
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=250)

# 6 Test
loss, acc = model.evaluate(X_test, y_test)
print(acc)
```

**5.DECISION TREE CLASSIFIER:--**

---

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score

df=load_iris()

x = df.data
y = df.target
print(x)
print(y)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=50, random_state=25)
print(x_train)
print(x_test)
print(y_train)
print(y_test)

from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)

y_train_pred = clf.predict(x_train)
y_test_pred = clf.predict(x_test)

print("Train data accuracy:", accuracy_score(y_train, y_train_pred))
print("Test data accuracy:", accuracy_score(y_test, y_test_pred))
```

**8.Clustering:--**

---

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/iris.csv').iloc[:, :-1].to_numpy()

# Elbow curve
plt.plot(range(1,11), [KMeans(k, n_init=10).fit(X).inertia_ for k in range(1,11)])
plt.show()

# KMeans (3 clusters)
kmeans = KMeans(3, n_init=10).fit(X)
labels = kmeans.labels_

# Plot clusters + centroids using only 2 features
for c in ['r','b','g']:
    i = ['r','b','g'].index(c)
    plt.scatter(X[labels==i,0], X[labels==i,1], c=c)

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            c='y', marker='*', s=150)
plt.show()
```

**9.K-MEANS**

---

```python
# 1️⃣ Import
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# 2️⃣ Load
df = pd.read_csv('/content/income.csv')

# 3️⃣ Prepare (scale Age and Income to 0–1 range)
scaler = MinMaxScaler()
df[['Age', 'Income($)']] = scaler.fit_transform(df[['Age', 'Income($)']])

# 4️⃣ Model (K-Means with 3 clusters)
kmeans = KMeans(n_clusters=3, random_state=0)
df['cluster'] = kmeans.fit_predict(df[['Age', 'Income($)']])

# 5️⃣ Visualize clusters & centroids
centers = kmeans.cluster_centers_

plt.scatter(df['Age'], df['Income($)'], c=df['cluster'])
plt.scatter(centers[:, 0], centers[:, 1],
            marker='*', s=200, label='centroids')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
plt.show()
```
